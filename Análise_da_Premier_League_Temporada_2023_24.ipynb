**Visão geral da temporada 2023/24 da Premier League**

A temporada 2023/24 marcou a 32ª edição da Premier League e a 125ª temporada do futebol de elite na Inglaterra. Com início em 11 de agosto de 2023 e término em 19 de maio de 2024, a campanha proporcionou momentos históricos e narrativas emocionantes.

O Manchester City defendeu com sucesso o seu título, tornando-se o primeiro clube na história da Premier League a conquistar quatro campeonatos consecutivos. Na outra ponta da tabela, todas as três equipes promovidas — Luton Town, Burnley e Sheffield United — foram rebaixadas, um fato raro que não se via desde a temporada 1997/98.

# **Importando bibliotecas e carregando dados**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
pd.options.mode.chained_assignment = None
```

```python
premierleague = pd.read_csv('premier-player-23-24.csv', sep = ',')
premierleague.head()
```

# **Variáveis**

*   Player: Nome do jogador.
*   Nation: Nacionalidade do jogador.
*   Pos: Posição do jogador (ex.: atacante, meio-campista, zagueiro).
*   Age: Idade do jogador.
*   MP (Minutos jogados): Total de minutos jogados pelo jogador.
*   Starts: Número de partidas em que o jogador foi titular.
*   Min (Minutos): Total de minutos jogados pelo jogador (este valor pode ser o mesmo que MP).
*   90s (90 minutos jogados): O equivalente a partidas de 90 minutos jogadas pelo jogador (ex.: 1,5 = 135 minutos).
*   Gls (Gols): Total de gols marcados pelo jogador.
*   Ast (Assistências): Total de assistências feitas pelo jogador.
*   G+A (Gols + Assistências): Total de gols e assistências combinados.
*   G-PK (Gols - Pênaltis): Total de gols marcados, excluindo pênaltis.
*   PK (Pênaltis): Número de gols de pênalti convertidos pelo jogador.
*   PKatt (Tentativas de pênalti): Número de pênaltis cobrados pelo jogador.
*   CrdY (Cartões amarelo): Número de cartões amarelos recebidos pelo jogador.
*   CrdR (Cartões vermelho): Número de cartões vermelhos recebidos pelo jogador.
*   xG (Expectativa de gols): Número esperado de gols a partir dos chutes do jogador.
*   npxG (Gols esperados sem pênalti): Gols esperados, excluindo pênaltis.
*   xAG (Expectativa de assistências): Número esperado de assistências a partir dos passes do jogador.
*   npxG+xAG (xG sem penalidades + xAG): Total de gols esperados (sem pênaltis) e assistências esperadas.
*   PrgC (Carregadas progressivas): Número de vezes que o jogador conduziu a bola para frente.
*   PrgP (Passes progressivos): Número de passes realizados pelo jogador que levaram a bola para frente.
*   PrgR (Corridas progressivas): Número de vezes que o jogador fez corridas para frente com a bola.
*   Gls (Gols): (Repetido, já definido) Total de gols marcados.
*   Ast (Assistências): (Repetido, já definido) Total de assistências realizadas.
*   G+A (Gols + Assistências): (Repetido, já definido) Total de gols e assistências combinados.
*   G-PK (Gols - Pênaltis): (Repetido, já definido) Gols marcados, excluindo pênaltis.
*   G+A-PK (Gols + Assistências - Pênaltis): Total de gols e assistências menos gols de pênalti.
*   xG (Expectativa de gols): (Repetido, já definido) Número esperado de gols a partir dos chutes do jogador.
*   xAG (Expectativa de assistências): (Repetido, já definido) Número esperado de assistências a partir dos passes do jogador.
*   xG+xAG (Expectativa de gols + Expectativa de assistências): Total de gols e assistências esperados.
*   npxG (Gols esperados sem pênalti): (Repetido, já definido) Gols esperados, excluindo pênaltis.
*   npxG+xAG (xG sem penalidades + xAG): Total de gols esperados (sem pênaltis) e assistências esperadas.

# **Análise por Clube**

```python
pl_clubs = premierleague[['Team', 'Gls', 'Ast', 'G+A', 'xG', 'xAG', '90s', 'PrgC', 'PrgP', 'PrgR', 'CrdY', 'CrdR']]
pl_clubs = pl_clubs.groupby('Team').agg({
    'Gls': 'sum',
    'Ast': 'sum',
    'G+A': 'sum',
    'xG': 'sum',
    'xAG': 'sum',
    '90s': 'sum',
    'PrgC': 'sum',
    'PrgP': 'sum',
    'PrgR': 'sum',
    'CrdY': 'sum',
    'CrdR': 'sum'
}).reset_index()
pl_clubs.sort_values('G+A', ascending = False, inplace = True)
pl_clubs
```

```python
figure, ax = plt.subplots(figsize = (10, 6))
g = ax.barh(pl_clubs['Team'], pl_clubs['Gls'], color = 'r', label = 'Goals')
a = ax.barh(pl_clubs['Team'], pl_clubs['Ast'], left = pl_clubs['Gls'], color = 'cyan', label = 'Ast')
ax.bar_label(g, label_type='center', color = 'white')
ax.bar_label(a, label_type='center')
ax.set_title('Total de gols + assistências por clube')
ax.legend()
plt.show()
```

Ao final da temporada, o Manchester City se destacou como o time com o ataque mais eficiente, marcando 94 gols e distribuindo 69 assistências — o melhor em ambas as estatísticas.

Por outro lado, o Sheffield United teve o menor número de gols marcados (31), enquanto o Burnley registrou o menor número de assistências (17).

```python
figure, ax = plt.subplots(figsize = (10, 6))

for i, row in pl_clubs.iterrows():
  plt.text(row['xG'] + 0.3, row['Gls'], row['Team'], fontsize=10)

cmap = plt.cm.get_cmap('tab20c', len(pl_clubs))
ax.scatter(pl_clubs['xG'], pl_clubs['Gls'], s = pl_clubs['Gls'] * 5, c = range(len(pl_clubs)), cmap = cmap, alpha = 0.7, label = 'Gols')
ax.set_title('Gols vs xG')
ax.set_xlabel('xG')
ax.set_ylabel('Gols')
plt.show()
```

Este gráfico de dispersão mostra a relação entre os gols reais marcados (Gls) e os gols esperados (xG) para cada equipe da Premier League na temporada 2023/24. A métrica de gols esperados (xG) avalia a qualidade das chances criadas, fornecendo informações sobre se uma equipe está finalizando com eficiência ou apresentando um desempenho abaixo do esperado.

Equipes acima da diagonal (onde Gols > xG) superaram as expectativas, provavelmente devido à finalização precisa ou a atuações individuais de destaque. Notavelmente, o Manchester City se destaca com o maior desempenho acima do esperado, marcando mais de 90 gols a partir de aproximadamente 85 xG. Arsenal e Newcastle United também superaram ligeiramente as expectativas.

Por outro lado, equipes abaixo da diagonal (onde Gols < xG) tiveram um desempenho abaixo do esperado em relação aos seus gols esperados, indicando ineficiências na conversão de chances. Por exemplo, Everton e Brighton criaram chances equivalentes a cerca de 50-55 xG, mas terminaram com menos gols, sugerindo oportunidades perdidas.

O Sheffield United, posicionado no canto inferior esquerdo, apresentou o menor xG e o menor número de gols, refletindo suas dificuldades ofensivas ao longo da temporada.

Essa comparação ajuda a revelar não apenas quantos gols as equipes marcaram, mas também a eficácia com que aproveitaram as oportunidades criadas.

```python
figure, ax = plt.subplots(figsize = (10, 6))

for i, row in pl_clubs.iterrows():
  plt.text(row['xAG'] + 0.3, row['Ast'], row['Team'], fontsize=10)

cmap = plt.cm.get_cmap('tab20c', len(pl_clubs))
ax.scatter(pl_clubs['xAG'], pl_clubs['Ast'], s = pl_clubs['Ast'] * 5, c = range(len(pl_clubs)), cmap = cmap, alpha = 0.7, label = 'Assists')
ax.set_title('Assistências vs xAG')
ax.set_xlabel('xAG')
ax.set_ylabel('Assistências')
plt.show()
```

Este gráfico de dispersão compara as assistências reais (Ast) com as assistências esperadas (xA) para cada equipe na temporada 2023/24 da Premier League. A métrica xA estima quantas assistências uma equipe deveria ter com base na qualidade de seus passes decisivos e chances criadas, independentemente de o chute final ter resultado em gol.

As equipes localizadas acima da linha diagonal (Ast > xA) superaram as expectativas, convertendo chances em assistências com mais eficiência. O Manchester City lidera novamente, com o maior desempenho acima do esperado nesta categoria — cerca de 65 assistências a partir de uma base de xA de aproximadamente 70. Arsenal, Tottenham e Liverpool também tiveram um bom desempenho, superando ligeiramente seus valores de assistências esperados.

As equipes abaixo da linha diagonal apresentaram desempenho inferior ao seu potencial criativo. Notavelmente, o Everton registrou significativamente menos assistências (17) em comparação com seu xA (40), sugerindo problemas em transformar boas chances em gols. Da mesma forma, o Sheffield United teve dificuldades tanto na criação quanto na conversão de oportunidades, terminando na última posição em ambas as métricas.

Este gráfico destaca quais equipes foram mais eficientes em transformar jogadas criativas em resultados ofensivos concretos — e quais equipes tiveram dificuldades em capitalizar em suas jogadas de construção de jogo.

```python
pl_clubs['TotalCrds'] = pl_clubs['CrdY'] + pl_clubs['CrdR']
pl_clubs.sort_values('TotalCrds', ascending = False, inplace = True)
```

```python
figure, ax = plt.subplots(figsize = (10, 6))
y = ax.barh(pl_clubs['Team'], pl_clubs['CrdY'], color = 'yellow')
r = ax.barh(pl_clubs['Team'], pl_clubs['CrdR'], left = pl_clubs['CrdY'], color = 'r')
ax.bar_label(y, label_type='center')
ax.bar_label(r, label_type='center')
ax.set_title('Total de cartões amarelos e vermelhos na temporada')
plt.show()
```

```python
pl_clubs['PrgTotal'] = pl_clubs['PrgC'] + pl_clubs['PrgP'] + pl_clubs['PrgR']
pl_clubs.sort_values('PrgTotal', ascending = False, inplace = True)
pl_clubs
```

```python
figure, ax = plt.subplots(figsize = (10, 6))
p = ax.barh(pl_clubs['Team'], pl_clubs['PrgTotal'], color = 'springgreen')
ax.set_title('Total de Progressão por clube')
ax.set_xlabel('Total de progressões')
ax.bar_label(p, label_type='edge')
plt.show()
```

Este gráfico de barras exibe o número total de ações de progressão de cada equipe da Premier League na temporada 2023/24. A métrica inclui uma contagem combinada de passes progressivos, condução de bola e corridas — ações que movem a bola significativamente mais perto do gol adversário e refletem a intenção ofensiva e o domínio territorial de uma equipe.

Como esperado, as equipes com perfis ofensivos mais fortes dominam o topo do gráfico. O Manchester City lidera a liga com 5.219 progressões totais, seguido de perto por Tottenham Hotspur, Liverpool e Arsenal — todos com mais de 5.000 ações de progressão. Essas equipes mantiveram altas taxas de posse de bola e foram consistentemente proativas no avanço da bola.

Em contraste, equipes na parte inferior do gráfico, como Sheffield United, Everton e Crystal Palace, tiveram significativamente menos ações de progressão. O Sheffield United, em particular, registrou apenas 2.101 progressões — o menor número da liga — reforçando suas dificuldades gerais na construção de jogo e na criação de oportunidades.

Essa métrica destaca a frequência com que cada equipe avançou com a bola para áreas mais perigosas, oferecendo uma visão mais ampla de seu estilo de jogo e atividade ofensiva, além de gols e assistências.

```python
positions = {
    'DF' : 'Zagueiro',
    'DF,FW' : 'Jogador polivalente',
    'DF,MF' : 'Lateral',
    'FW' : 'Atacante',
    'FW,DF' : 'Polivalente',
    'FW,MF' : 'Ponta',
    'GK' : 'Goleiro',
    'MF' : 'Meio campista',
    'MF,DF' : 'Volante',
    'MF,FW' : 'Meia atacante',
}

pl_pos['Position'] = pl_pos['Pos'].map(positions)
pl_pos
```

```python
pl_pos = pl_pos.groupby('Position').sum().reset_index()
pl_pos.drop(columns = ['Pos'], inplace = True)
pl_pos
```

```python
pl_pos['PrgTotal'] = pl_pos['PrgC'] + pl_pos['PrgP'] + pl_pos['PrgR']
pl_pos['TotalCrd'] = pl_pos['CrdY'] + pl_pos['CrdR']
pl_pos
```

```python
pl_pos.sort_values('Player', ascending = False, inplace = True)

figure, ax = plt.subplots(figsize = (10, 6))
p = ax.bar(pl_pos['Position'], pl_pos['Player'], color = 'cyan')
ax.set_title('Total de jogadores por posição')
ax.bar_label(p, label_type='edge')
plt.xlabel('Posições')
plt.xticks(rotation = 45)
plt.show()
```

```python
pl_pos.sort_values('Gls', ascending = False, inplace = True)

figure, ax = plt.subplots(figsize = (10, 6))
p = ax.bar(pl_pos['Position'], pl_pos['Gls'], color = 'Red')
ax.set_title('Total de gols por posição')
ax.bar_label(p, label_type='edge')
plt.xlabel('Posições')
plt.xticks(rotation = 45)
plt.show()
```

```python
pl_pos.sort_values('G+A', ascending = False, inplace = True)

figure, ax = plt.subplots(figsize = (10, 6))
p = ax.bar(pl_pos['Position'], pl_pos['G+A'], color = 'springgreen')
ax.set_title('Total de participação em gols por posição')
ax.bar_label(p, label_type='edge')
plt.xlabel('Posições')
plt.xticks(rotation = 45)
plt.show()
```

```python
pl_pos.sort_values('CrdY', ascending = False, inplace = True)

figure, ax = plt.subplots(figsize = (10, 6))
y = ax.bar(pl_pos['Position'], pl_pos['CrdY'], color = 'yellow')
r = ax.bar(pl_pos['Position'], pl_pos['CrdR'], bottom = pl_pos['CrdY'], color = 'r')
ax.set_title('Total de cartões por posição')
ax.bar_label(y, label_type='center')
ax.bar_label(r, label_type='center')
plt.xlabel('Posições')
plt.xticks(rotation = 45)
plt.show()
```

```python
pl_pos.sort_values('PrgTotal', ascending = False, inplace = True)

figure, ax = plt.subplots(figsize = (10, 6))
c = ax.bar(pl_pos['Position'], pl_pos['PrgC'], color = 'g', label = 'Carregada progressiva')
p = ax.bar(pl_pos['Position'], pl_pos['PrgP'], bottom = pl_pos['PrgC'], color = 'b', label = 'Passes progressivos')
r = ax.bar(pl_pos['Position'], pl_pos['PrgR'], bottom = pl_pos['PrgC'] + pl_pos['PrgP'], color = 'r', label = 'Corridas progressivas')
ax.set_title('Total de carregadas por posição')
ax.bar_label(c, label_type='center')
ax.bar_label(p, label_type='center')
ax.bar_label(r, label_type='center')
plt.xlabel('Posições')
plt.xticks(rotation = 45)
ax.legend()
plt.show()
```

Os dados revelam que os defensores dominam tanto a condução quanto a progressão de passes. Isso está de acordo com seu papel tático, no qual são responsáveis ​​por avançar com a bola, seja por meio de passes para meio-campistas ou pontas, iniciando assim transições ofensivas. Em contrapartida, os atacantes demonstram superioridade na progressão com dribles, refletindo sua função principal de conduzir a bola para o terço ofensivo por meio de corridas individuais. Essa distinção destaca os papéis complementares dentro da equipe, com os defensores focando na progressão da bola por meio da distribuição e os atacantes explorando o espaço com condução direta.
